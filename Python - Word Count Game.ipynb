# >>>>>>>>>>>>>>>>> **Word Count Game** <<<<<<<<<<<<<<<<


### Regras: *2 jogadores são solicitados a estimarem quantas vezes determinada palavra aparece em uma música. A resposta mais próxima da correta ganha a rodada.* 

### Rules: *2 players are asked to guess how many times a word is in a song. The closest answer to the correct one wins the round.*
